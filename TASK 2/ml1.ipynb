```python

df.columns = df.columns.str.strip()  


y = df["CO2 Emissions(g/km)"]
X = df[["Engine Size(L)", "Cylinders", "Fuel Consumption Comb (L/100 km)"]]
```

```python
X = (X - X.mean()) / X.std()
y = (y - y.mean()) / y.std()
```

```python
# 80/20 split
split_idx = int(0.8 * len(X))
X_train, X_test = X.iloc[:split_idx], X.iloc[split_idx:]
y_train, y_test = y.iloc[:split_idx], y.iloc[split_idx:]

# Add bias column
X_train = np.c_[np.ones(X_train.shape[0]), X_train]
X_test = np.c_[np.ones(X_test.shape[0]), X_test]
```

```python
# Convert to numpy
X_train_np = np.array(X_train)
y_train_np = np.array(y_train)

#  = (XX)Xy
theta = np.linalg.inv(X_train_np.T @ X_train_np) @ X_train_np.T @ y_train_np
```

```python
y_pred_train = X_train_np @ theta
y_pred_test = np.array(X_test) @ theta
```

```python
def r2_score(y_true, y_pred):
    ss_total = np.sum((y_true - y_true.mean())**2)
    ss_res = np.sum((y_true - y_pred)**2)
    return 1 - ss_res / ss_total

def rmse(y_true, y_pred):
    return np.sqrt(np.mean((y_true - y_pred)**2))

print("R (Train):", r2_score(y_train, y_pred_train))
print("R (Test):", r2_score(y_test, y_pred_test))
print("RMSE (Test):", rmse(y_test, y_pred_test))
```

```python
plt.figure(figsize=(8,5))
plt.scatter(y_test, y_pred_test, alpha=0.5)
plt.plot([-3, 3], [-3, 3], color='red')  # ideal line
plt.title("Actual vs Predicted CO2 Emissions (Standardized)")
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.grid(True)
plt.show()
```

```python
# Initialize parameters
theta = np.zeros(X_train.shape[1])
learning_rate = 0.01
epochs = 1000

losses = []

for epoch in range(epochs):
    # Predictions
    y_pred = X_train @ theta

    # Error
    error = y_pred - y_train

    # Gradient of MSE w.r.t theta:  = (2/n) * X(X - y)
    gradient = (2 / len(X_train)) * (X_train.T @ error)

    # Update weights
    theta -= learning_rate * gradient

    # Compute loss (MSE)
    mse = np.mean(error ** 2)
    losses.append(mse)

    # Print every 100 epochs
    if epoch % 100 == 0:
        print(f"Epoch {epoch}: MSE = {mse:.4f}")
```

```python
plt.figure(figsize=(8,5))
plt.plot(losses)
plt.title("Loss Curve (MSE over Epochs)")
plt.xlabel("Epoch")
plt.ylabel("MSE")
plt.grid(True)
plt.show()
```
