```python
# IPL Deliveries Dataset - Exploratory Data Analysis (EDA)
# This notebook provides a comprehensive EDA of the IPL deliveries dataset, including summary statistics and a variety of visualizations to understand the data better.
# 1. Import Required Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set(style='whitegrid')
%matplotlib inline
# 2. Load the Dataset

df = pd.read_csv('deliveries.csv')
df.head()
# 3. View Dataset Structure and Basic Info

display(df.head())
print('Shape:', df.shape)
print('Columns:', df.columns.tolist())
print('\nData Types:')
print(df.dtypes)
# 4. Check for Missing Values

missing = df.isnull().sum()
print(missing)
plt.figure(figsize=(12,6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()
## 5. Summary Statistics of Numeric Columns

df.describe().T
## 6. Distribution of Runs per Ball

plt.figure(figsize=(8,5))
sns.countplot(x='total_runs', data=df, palette='Blues')
plt.title('Distribution of Total Runs per Delivery')
plt.xlabel('Total Runs (per ball)')
plt.ylabel('Count')
plt.show()
## 7. Top Batsmen by Total Runs

top_batsmen = df.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_batsmen.values, y=top_batsmen.index, palette='mako')
plt.title('Top 10 Batsmen by Total Runs')
plt.xlabel('Total Runs')
plt.ylabel('Batsman')
plt.show()
## 8. Top Bowlers by Wickets Taken

# Exclude run outs and retired hurt from wickets
dismissal_exclude = ['run out', 'retired hurt', 'obstructing the field']
wickets = df[~df['dismissal_kind'].isin(dismissal_exclude) & df['dismissal_kind'].notnull()]
top_bowlers = wickets['bowler'].value_counts().head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_bowlers.values, y=top_bowlers.index, palette='crest')
plt.title('Top 10 Bowlers by Wickets Taken')
plt.xlabel('Wickets')
plt.ylabel('Bowler')
plt.show()
## 9. Dismissal Types Distribution

plt.figure(figsize=(10,5))
sns.countplot(y='dismissal_kind', data=df[df['dismissal_kind'].notnull()], order=df['dismissal_kind'].value_counts().index, palette='flare')
plt.title('Distribution of Dismissal Types')
plt.xlabel('Count')
plt.ylabel('Dismissal Kind')
plt.show()
## 10. Extras Analysis (Wides, No Balls, Leg Byes, Byes, Penalties)

extras = ['wide_runs', 'noball_runs', 'legbye_runs', 'bye_runs', 'penalty_runs']
extras_df = df[extras].sum().reset_index()
extras_df.columns = ['Extra Type', 'Total']
plt.figure(figsize=(8,5))
sns.barplot(x='Extra Type', y='Total', data=extras_df, palette='Set2')
plt.title('Total Extras by Type')
plt.ylabel('Total Extras')
plt.show()
## 11. Runs per Over Analysis

over_runs = df.groupby('over')['total_runs'].mean()
plt.figure(figsize=(10,5))
sns.lineplot(x=over_runs.index, y=over_runs.values, marker='o')
plt.title('Average Runs per Over')
plt.xlabel('Over')
plt.ylabel('Average Runs')
plt.show()
## 12. Team-wise Total Runs

team_runs = df.groupby('batting_team')['total_runs'].sum().sort_values(ascending=False)
plt.figure(figsize=(12,6))
sns.barplot(x=team_runs.values, y=team_runs.index, palette='viridis')
plt.title('Total Runs by Team')
plt.xlabel('Total Runs')
plt.ylabel('Team')
plt.show()
## 13. Inning-wise Comparison

inning_runs = df.groupby('inning')['total_runs'].sum()
plt.figure(figsize=(7,5))
sns.barplot(x=inning_runs.index, y=inning_runs.values, palette='coolwarm')
plt.title('Total Runs: 1st vs 2nd Innings')
plt.xlabel('Inning')
plt.ylabel('Total Runs')
plt.show()

plt.figure(figsize=(7,5))
sns.boxplot(x='inning', y='total_runs', data=df, palette='coolwarm')
plt.title('Distribution of Runs per Ball by Inning')
plt.xlabel('Inning')
plt.ylabel('Runs per Ball')
plt.show()
## 14. Correlation Heatmap of Numeric Features

plt.figure(figsize=(10,8))
corr = df.select_dtypes(include=np.number).corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Numeric Features')
plt.show()
```
